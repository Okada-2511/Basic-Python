# Chapter 2.2 - Exercise 2: Hãy thực hiện những yêu cầu liên quan tới Data Frame

```python
import numpy as np
import pandas as pd
```

```python
# Câu 1: Cho dictionary như sau:
dic_1 = {'X':[78,85,96,80,86], 'Y':[84,94,89,83,86], 'Z':[86,97,96,72,83]}

# Tạo dataframe df1 từ dic_1
df1 = pd.DataFrame(dic_1)
# In nội dung của dataframe df1
print(df1)
```

```python
# Câu 2:
# Cho dictionary exam_data như sau:
exam_data  = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5,np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

# Cho list labels như sau:
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
# Câu 2a: Tạo dataframe df2 từ exam_data, với index của dataframe là lables
df2 = pd.DataFrame(exam_data,index = labels)

# In nội dung của dataframe df2
print(df2)
```

```python
# Câu 2b: Xem thông tin (info()) của dataframe df2

print(df2.info())
```

```python
# Câu 3: Tạo dataframe df3 từ df2, chỉ chứa 2 cột là name và score
df3= df2.loc[:,['name','score']]
# Xem kiểu dữ liệu (type) và kích thước (shape) của df3
print(type(df3))
print(df3.shape)
# Hiển thị các dòng dữ liệu đầu tiên (head) của df3
print(df3.head())
```

```python
# Câu 4: Hiển thị thông tin thống kê chung (describe) của dataframe df2
print(df3.describe(include='all'))
```

```python
# Câu 5: Tạo dataframe df4 từ df2, chỉ chứa 2 cột là name và score, và chỉ có các dòng 1, 3, 5, 6
df4= df2.iloc[[1,3,5,6],[0,1]]
# In nội dung của dataframe df4
print(df4)
```

```python
# Câu 6: Từ dataframe df2, in các dòng có dữ liệu score bị null
df2.loc[df2['score'].isnull() ,: ]
```

```python
# Câu 7: Từ dataframe df2, in các dòng có score > 15 và <=20
df2.loc[ (df2['score'] > 15) & (df2['score'] <= 20) ,:]
```

```python
# Câu 8: Cập nhật điểm (score) ở dòng 'd' thành 18
df2.loc['d','score'] = 18
df2
```

```python
# Câu 9: Cho biết điểm (score) nào có tần suất xuất hiện nhiều nhất trong df2, và in ra những dòng có điểm là tần suất xuất hiện nhiều nhất
df2.loc[:,'score'].mode()
df2.loc[df2['score'] == 9 ,:]

#c2
df2['score'].value_counts()
```

```python
# Câu 10: Thêm dòng k có nội dung như sau: ['Suresh', 15.5, 1, 'yes'] vào df2
df2.loc['k'] = ['Suresh', 15.5, 1, 'yes']
# Hiển thị 5 dòng cuối cùng (tail) của df2.
print(df2.tail(5))
```

```python
# Câu 11a: Thêm dòng l có nội dung như sau: ['Janny', 12.5, 2, 'yes'] vào df2.
df2.loc['l'] = ['Janny', 12.5, 2, 'yes']
# Hiển thị 5 dòng cuối cùng (tail) của df2.
df2.tail(5)
```

```python
# Câu 11b: Xóa bỏ dòng l của df2.

#df2 = df2.drop(['l'])
# Hiển thị lại 5 dòng cuối cùng (tail) của df2.
df2.tail(5)
```

```python
# Câu 12: Sắp xếp df2 tăng dần theo điểm (score)
df2.sort_values(by = 'score')
```

```python
# Câu 13: Thêm cột result vào df2,
# dựa vào dữ liệu của cột'score', nếu dòng nào có điểm >=10 thì giá trị của cột result = 1, ngược lại = 0
def result(score):
   if score >= 10:
    return 1
   else:
      return 0
df2['result'] = df2['score'].map(result)
# In nội dung của dataframe df2
df2
```

```python
# Câu 14: Trong df2, thay tên 'Emily' thành 'Samantha'
df2.loc[ df2['name'] == 'Emily','name' ]= 'Samantha'
# In lại nội dung của dataframe df2
df2
```
