```python
# Câu 2: Vẽ 2 biều đồ line plot ở câu 1 nhưng trên 2 vùng:
# vùng 1: plt.axes([0.05, 0.05, 0.8, 0.8]), và vùng 2: plt.axes([0.9, 0.05, 0.8, 0.8])

# Tạo một figure mới
plt.figure()

# Tạo ax1 với vị trí và kích thước được chỉ định
# [left, bottom, width, height]
# 5% từ trái, 5% từ dưới, rộng 80%, cao 80%
ax1 = plt.axes([0.05, 0.05, 0.8, 0.8])

# Vẽ biểu đồ trên ax1
ax1.plot(year, physical_sciences, color='blue')
ax1.set_title('Ty le phan tram bang khoa hoc vat ly duoc tra cho phu nu qua cac nam')
ax1.set_xlabel('Nam')
ax1.set_ylabel('Ty le phan tram')
# Đặt ax2 ở vị trí khác để không chồng lấn

ax2 = plt.axes([0.9, 0.05, 0.8, 0.8])
ax2.plot(year, computer_sciences, color='red')
ax2.set_title('Ty le phan tram bang khoa hoc may tinh duoc tra cho phu nu qua cac nam')
ax2.set_xlabel('Nam')
ax2.set_ylabel('Ty le phan tram')

plt.show()
```

```python
# Câu 3: Vẽ 2 biểu đồ line plot ở câu 1 nhưng trên 2 subplot:
# subplot 1: plt.subplot(1, 2, 1), và subplot 2: plt.subplot(1, 2, 2),
# Lưu ý: sử dụng plt.tight_layout() trước khi show()
plt.figure(figsize=(12, 6))
# Vẽ biểu đồ line plot cho subplot 1
ax1 = plt.subplot(1, 2, 1)
ax1.plot(year, physical_sciences, color='blue')
# Vẽ biểu đồ line plot cho subplot 2
ax2 = plt.subplot(1, 2, 2)
ax2.plot(year, computer_sciences, color='red')

# Sử dụng plt.tight_layout() trước khi show()
plt.tight_layout()
plt.show()

```

```python
# Câu 4: Vẽ 4 biểu đồ line với 4 màu khác nhau:
# Biểu đồ 1: thể hiện tỷ lệ % bằng Khoa học vật lý được trao cho phụ nữ qua các năm (year - physical_sciences)
# Biểu đồ 2: thể hiện tỷ lệ % bằng Khoa học máy tính được trao cho phụ nữ qua các năm (year - computer_science)
# Biểu đồ 3: thể hiện tỷ lệ %  phụ nữ tham gia các công việc liên quan đến y tế qua các năm (year - health)
# Biểu đồ 4: thể hiện tỷ lệ %  phụ nữ tham gia các công việc liên quan đến giáo dục qua các năm (year - education)

# trên 4 subplot: plt.subplot(2, 2, 1), plt.subplot(2, 2, 2), plt.subplot(2, 2, 3), plt.subplot(2, 2, 4)
plt.figure()
## Vẽ biểu đồ line cho line 1
ax1 = plt.subplot(2, 2, 1)
ax1.plot(year, physical_sciences, color='blue')

## Vẽ biểu đồ line cho line 2
ax2 = plt.subplot(2, 2, 2)
ax2.plot(year, computer_sciences, color='red')

## Vẽ biểu đồ line cho line 3
ax3 = plt.subplot(2, 2, 3)
ax3.plot(year, health, color='green')

## Vẽ biểu đồ line cho line 4
ax4 = plt.subplot(2, 2, 4)
ax4.plot(year, education, color='orange')

## Sử dụng plt.tight_layout() trước khi show()
plt.tight_layout()
plt.show()
```

```python
# Câu 5: Vẽ 2 biểu đồ line plot:
# Biểu đồ 1: thể hiện tỷ lệ % bằng Khoa học máy tính được trao cho phụ nữ qua các năm (year - computer_science)
# Biểu đồ 2: thể hiện tỷ lệ % bằng Khoa học vật lý được trao cho phụ nữ qua các năm (year - physical_sciences)
# nhưng giới hạn từ năm 1980-2000 --> có plt.xlim(1980, 2000) và plt.ylim(0, 50)
plt.plot(year, computer_sciences, color='red')
plt.plot(year, physical_sciences, color = 'blue')

# Đặt nhãn cho các axis
plt.title('Ty le phan tram bang khoa hoc may tinh duoc tra cho phu nu qua cac nam')
plt.xlabel('Nam')
plt.ylabel('Ty le phan tram')
plt.legend(['Khoa hoc may tinh', 'Khoa hoc vat ly'])

# Đặt giới hạn cho x-axis và y-axis
plt.xlim(1980,2000)
plt.ylim(0,50)

# Đặt tiêu đề và hiển thị biểu đồ
plt.show()
```
