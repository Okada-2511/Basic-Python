#### Dựa trên tập tin _train.csv_ , hãy thực hiện các yêu cầu sau:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Câu 1:
# a) Đọc dữ liệu từ tập tin train.csv và lưu vào biến titanic.
df_titanic = pd.read_csv('train.csv')
# Hiển thị 5 dòng dữ liệu đầu của titanic
df_titanic.head(5)
```

```python
# b) Thiết lập cột index cho tatanic là PassengerId. Hiển thị lại 5 dòng dữ liệu đầu của titanic lúc này.
df_titanic.set_index('PassengerId', inplace=True)
df_titanic.head(5)
```

```python
# Câu 2: Tạo pie chart thể hiện tỷ lệ hành khách nam/nữ trên tàu.

# Tạo biến males, female là tổng nam và tổng nữ
males = df_titanic.loc[ df_titanic['Sex']== 'male' ,'Sex'].count()
female = df_titanic.loc[df_titanic['Sex']== 'female' ,'Sex'].count()
#print(df_titanic['Sex'].unique())
# Tạo biến proportions là list có 2 phần tử là male và female
proportions = [males, female]
# Vẽ biểu đồ
plt.pie(proportions,labels= ['Male','Female'], colors =['blue','red'],autopct='%.2f%%',startangle=90,explode=(0,0.1))

# Thiết lập title là Sex Proportion
plt.title('Ty le nam/nu')

# Show the plot
plt.show()
```

```python
#ve bieu do pie tinh ty le song sot giuwa nam va nu
males_survived = df_titanic.loc[(df_titanic['Sex']=='male') & (df_titanic['Survived'] == 1) ,'Sex'].count()
female_survived = df_titanic.loc[(df_titanic['Sex']=='female') & (df_titanic['Survived'] == 1) ,'Sex'].count()
survived_proportions = [males_survived, female_survived]
plt.pie(survived_proportions,labels=['males_survived','female_survived'],colors=['blue','red'],autopct='%2.f%%',startangle=90,explode=(0,0.1))
plt.title('Ty le nam/nu song sot')
plt.show()
```

```python
# Câu 3: Cho biết có bao nhiêu người còn sống sót
df_titanic.loc[ df_titanic['Survived']==1,'Survived'].count()
```

```python
# Câu 4: Vẽ biểu đồ histogram của cột vé (Fare)

# Sắp xếp lại dữ liệu df theo cột Fare giảm dần, và hiển thị các dòng dữ liệu đầu của df
df_titanic.sort_values(by='Fare',ascending= False ,inplace = True)

# Tạo binsVal = np.arange(0,600,10)

binsVal = np.arange(0,600,10)
# Vẽ histogram với dữ liệu là df, bins = binsVal
plt.hist(df_titanic['Fare'],bins=binsVal)

# Thiết lập các thông tin xlabel, ylabel, title
plt.title('Bieu do tan xuat cua dac trung Fare')
plt.xlabel('Fare')
plt.ylabel('Tan xuat')

# Show the plot
plt.show()
```
